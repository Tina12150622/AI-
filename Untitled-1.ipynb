```python
import pandas as pd
import numpy as np
from scipy.stats.mstats import winsorize

# è¯»å–æ•°æ®
df = pd.read_csv("C:/Users/lemon/Desktop/AIèµ‹èƒ½/CNDD-0299 ä¸Šå¸‚å…¬å¸äººå·¥æ™ºèƒ½æŠ•èµ„æ°´å¹³æ–°.csv")

print(f"åŸå§‹æ ·æœ¬æ•°: {len(df)}")
print("æ•°æ®åˆ—å:", df.columns.tolist())

# 1. å‰”é™¤ ST/*ST å…¬å¸
# ä½¿ç”¨ ~ è€Œä¸æ˜¯ -ï¼Œå¹¶å¤„ç†å¯èƒ½çš„NaNå€¼
df = df[~df["shortname"].str.contains("ST", na=False)]

# 2. æ£€æŸ¥å®é™…å­˜åœ¨çš„å…³é”®å˜é‡
# å…ˆæŸ¥çœ‹æœ‰å“ªäº›AIç›¸å…³çš„åˆ—
ai_columns = [col for col in df.columns if any(keyword in col.lower() for keyword in ['ai', 'invest'])]
print(f"æ‰¾åˆ°çš„AIç›¸å…³åˆ—: {ai_columns}")

# æ ¹æ®å¸¸è§çš„åˆ—åçŒœæµ‹ï¼Œæˆ–è€…æ‰‹åŠ¨æŒ‡å®š
# å¦‚æœåˆ—åä¸åŒï¼Œè¯·æ ¹æ®å®é™…æƒ…å†µä¿®æ”¹ä¸‹é¢çš„åˆ—è¡¨
key_columns = []
possible_names = ['aisoftinvest', 'aihardinvest', 'ainwesttotal', 'ainwestlevel', 
                  'ai_invest', 'ai_investment', 'ai_total', 'ai_level']

for name in possible_names:
    if name in df.columns:
        key_columns.append(name)
        print(f"æ‰¾åˆ°åˆ—: {name}")

if not key_columns:
    print("æœªæ‰¾åˆ°é¢„æœŸçš„AIæŠ•èµ„åˆ—ï¼Œè¯·æ‰‹åŠ¨æŒ‡å®šåˆ—å")
    # å¦‚æœè‡ªåŠ¨æŸ¥æ‰¾å¤±è´¥ï¼Œè¯·åœ¨è¿™é‡Œæ‰‹åŠ¨æŒ‡å®šåˆ—å
    # key_columns = ['ä½ çš„å®é™…åˆ—å1', 'ä½ çš„å®é™…åˆ—å2']
else:
    print(f"å°†ä½¿ç”¨ä»¥ä¸‹åˆ—è¿›è¡Œæ¸…æ´—: {key_columns}")

# å‰”é™¤å…³é”®å˜é‡ç¼ºå¤±çš„æ ·æœ¬
df_clean = df.dropna(subset=key_columns)

# 3. å¯¹è¿ç»­å˜é‡è¿›è¡Œ winsorize å¤„ç†
continuous_vars = key_columns  # ä½¿ç”¨å®é™…æ‰¾åˆ°çš„åˆ—
for col in continuous_vars:
    if col in df_clean.columns:
        # ç¡®ä¿æ•°æ®æ˜¯æ•°å€¼å‹
        df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')
        # è¿›è¡Œwinsorizeå¤„ç†
        df_clean[col] = winsorize(df_clean[col], limits=[0.01, 0.01])
        print(f"å·²å¤„ç†åˆ—: {col}")

# æŸ¥çœ‹å¤„ç†åçš„æ•°æ®
print(f"\næ¸…æ´—åæ ·æœ¬æ•°: {len(df_clean)}")
print("å¤„ç†åçš„æ•°æ®å‰5è¡Œ:")
print(df_clean[continuous_vars].head())

# ä¿å­˜æ¸…æ´—åçš„æ•°æ®
df_clean.to_csv("C:/Users/lemon/Desktop/AIèµ‹èƒ½/cleaned_AI_data.csv", 
                index=False, encoding='utf-8-sig')
print("âœ… æ¸…æ´—åçš„æ•°æ®å·²ä¿å­˜!")
```

```python
import pandas as pd
import numpy as np
from scipy.stats.mstats import winsorize

# è¯»å–æ¸…æ´—åçš„æ•°æ®
df = pd.read_csv("C:/Users/lemon/Desktop/AIèµ‹èƒ½/cleaned_AI_data.csv")

print(f"åˆå§‹æ ·æœ¬æ•°: {len(df)}")
print(f"æ•°æ®åˆ—å: {df.columns.tolist()}")

# ==================== æ„é€ AIæŠ•èµ„å˜é‡ ====================
print("æ­£åœ¨æ„é€ AIæŠ•èµ„å˜é‡...")

# ä½¿ç”¨æ•°æ®ä¸­å®é™…å­˜åœ¨çš„åˆ—å
df['AI_Intensity'] = df['aiinvesttotal']
df['AI_Amount'] = np.log(1 + df['aiinvesttotal'])
df['AI_Dummy'] = (df['aiinvesttotal'] > 0).astype(int)

print(f"AIæŠ•èµ„å˜é‡æ„é€ å®Œæˆï¼Œæœ‰AIæŠ•èµ„çš„å…¬å¸-å¹´åº¦è§‚æµ‹æ•°: {df['AI_Dummy'].sum()}")

# ==================== æ„é€ PSM-DIDå˜é‡ ====================
print("æ­£åœ¨æ„é€ PSM-DIDå˜é‡...")

# æŒ‰æ­£ç¡®çš„åˆ—åæ’åº
df = df.sort_values(['stkcd', 'year'])

# è¯†åˆ«é¦–æ¬¡AIæŠ•èµ„
first_ai_df = df[df['AI_Dummy'] == 1].groupby('stkcd')['year'].min().reset_index()
first_ai_df.columns = ['stkcd', 'first_ai_year']

df = df.merge(first_ai_df, on='stkcd', how='left')

# æ„é€ å¤„ç†å˜é‡
df['Treat'] = (~df['first_ai_year'].isna()).astype(int)
df['Post'] = (df['year'] >= df['first_ai_year']).astype(int)
df['Post'] = df['Post'].fillna(0)

print(f"å¤„ç†ç»„å…¬å¸æ•°: {df['Treat'].sum()}")

# ==================== æ„é€ å·¥å…·å˜é‡ ====================
print("æ­£åœ¨æ„é€ å·¥å…·å˜é‡...")

# IV1: è¡Œä¸š-å¹´åº¦å¹³å‡AIæŠ•èµ„æ°´å¹³
if 'industry' in df.columns:
    df['IV1'] = df.groupby(['industry', 'year'])['AI_Intensity'].transform('mean')
else:
    df['IV1'] = df.groupby('year')['AI_Intensity'].transform('mean')

# IV2: å ä½ç¬¦ï¼ˆéœ€è¦å¤–éƒ¨æ•°æ®ï¼‰
df['IV2'] = np.nan

# ==================== æ•°æ®é¢„å¤„ç† ====================
print("æ­£åœ¨è¿›è¡Œæ•°æ®é¢„å¤„ç†...")

# ç¼©å°¾å¤„ç†
continuous_vars = ['AI_Intensity', 'AI_Amount', 'aisoftinvest', 'aihardinvest', 'aiinvesttotal', 'aiinvestlevel']

for var in continuous_vars:
    if var in df.columns:
        df[var] = winsorize(df[var], limits=[0.01, 0.01])

# åˆ›å»ºæ»åå˜é‡
lag_vars = ['AI_Intensity', 'AI_Amount', 'AI_Dummy']
for var in lag_vars:
    if var in df.columns:
        df[f'{var}_lag1'] = df.groupby('stkcd')[var].shift(1)

# ==================== ä¿å­˜æœ€ç»ˆæ•°æ® ====================
print("æ­£åœ¨ä¿å­˜æ•°æ®...")

# åˆ é™¤å…³é”®å˜é‡ç¼ºå¤±çš„è§‚æµ‹å€¼
required_vars = ['stkcd', 'year', 'AI_Intensity']
df_final = df.dropna(subset=required_vars)

print(f"æœ€ç»ˆæœ‰æ•ˆæ ·æœ¬æ•°: {len(df_final)}")
print(f"æœ€ç»ˆå…¬å¸æ•°: {df_final['stkcd'].nunique()}")

# ä¿å­˜æ•°æ®
output_path = "C:/Users/lemon/Desktop/AIèµ‹èƒ½/final_analysis_data.csv"
df_final.to_csv(output_path, index=False, encoding='utf-8-sig')

print("âœ… æ•°æ®å¤„ç†å®Œæˆï¼")
print(f"ğŸ“ æ–‡ä»¶å·²ä¿å­˜è‡³: {output_path}")

# æ•°æ®æ¦‚è§ˆ
print("\n=== æ•°æ®æ¦‚è§ˆ ===")
print(f"æ—¶é—´èŒƒå›´: {df_final['year'].min()} - {df_final['year'].max()}")
print(f"æ€»è§‚æµ‹æ•°: {len(df_final)}")
print(f"AIæŠ•èµ„è§‚æµ‹æ•°: {df_final['AI_Dummy'].sum()}")
print(f"AIæŠ•èµ„å¼ºåº¦ç»Ÿè®¡:")
print(df_final['AI_Intensity'].describe())
```
